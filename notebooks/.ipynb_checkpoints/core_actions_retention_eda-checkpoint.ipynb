# Core actions vs churn proxy (EDA)

This notebook reproduces the exploratory analysis:

- **Universe**: subscribers (`first_subscription_date` present)
- **Outcome (proxy)**: `cancel_date` present vs absent
- **Adoption**: for each `action_type_id` a, `sum(total_usage) > 0` for that customer/action

**Important**: association / hypothesis generation only (no event timestamps in `intuit_usage.csv`).

## Inputs
- `../customers.csv`
- `../intuit_usage.csv`

## How to run

````python
import csv
import warnings
from collections import defaultdict
from datetime import date

import numpy as np
from sklearn.exceptions import ConvergenceWarning
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_auc_score
from sklearn.preprocessing import OneHotEncoder, StandardScaler

warnings.filterwarnings("ignore", category=ConvergenceWarning)
warnings.filterwarnings("ignore", category=FutureWarning)

CUSTOMERS_PATH = "../customers.csv"
USAGE_PATH = "../intuit_usage.csv"

# Crude "as-of" date for tenure feature (UTC calendar date).
SNAPSHOT_DATE = date(2026, 4, 11)
````

````python
def load_subscribers(path):
    subscriber_ids = set()
    churned_ids = set()
    channel_by_id = {}
    signup_by_id = {}

    with open(path, newline="") as f:
        r = csv.reader(f)
        next(r)  # header
        for row in r:
            if not row or len(row) < 5:
                continue
            cid = int(row[0])
            signup = row[1].strip()
            channel = row[2].strip() or "Unknown"
            first_sub = row[3].strip()
            cancel = row[4].strip()

            if not first_sub:
                continue

            subscriber_ids.add(cid)
            channel_by_id[cid] = channel
            signup_by_id[cid] = date.fromisoformat(signup)
            if cancel:
                churned_ids.add(cid)

    return subscriber_ids, churned_ids, channel_by_id, signup_by_id


def load_usage_sums(path, subscriber_ids):
    usage_sum = defaultdict(int)
    with open(path, newline="") as f:
        dr = csv.DictReader(f)
        for row in dr:
            cid = int(row["CUSTOMERID"])
            if cid not in subscriber_ids:
                continue
            act = int(row["action_type_id"])
            usage_sum[(cid, act)] += int(row["total_usage"])
    return usage_sum


subscriber_ids, churned_ids, channel_by_id, signup_by_id = load_subscribers(CUSTOMERS_PATH)
not_churned_ids = subscriber_ids - churned_ids

usage_sum = load_usage_sums(USAGE_PATH, subscriber_ids)
actions = sorted({act for (_, act) in usage_sum.keys()})

print(f"subscribers={len(subscriber_ids):,}")
print(f"churned_proxy(cancel_date_present)={len(churned_ids):,}")
print(f"not_churned_proxy={len(not_churned_ids):,}")
print(f"actions_observed={actions}")
````

## Lift table (bivariate)

\[
\text{lift}_a = P(\text{adopt}\mid \text{not churned}) - P(\text{adopt}\mid \text{churned})
\]


````python
n_ch = len(churned_ids)
n_nc = len(not_churned_ids)

rows = []
for act in actions:
    adopted_churn = sum(1 for cid in churned_ids if usage_sum.get((cid, act), 0) > 0)
    adopted_not = sum(1 for cid in not_churned_ids if usage_sum.get((cid, act), 0) > 0)

    p_ch = adopted_churn / n_ch
    p_nc = adopted_not / n_nc
    lift = p_nc - p_ch

    rows.append((act, adopted_churn, adopted_not, p_ch, p_nc, lift))

rows.sort(key=lambda r: r[5], reverse=True)

print("action\tadopt_c\tadopt_nc\tp_ch\tp_nc\tlift")
for act, ac, an, p_ch, p_nc, lift in rows:
    print(f"{act}\t{ac}\t{an}\t{p_ch:.4f}\t{p_nc:.4f}\t{lift:+.4f}")
````

## Multivariate logistic regression (exploratory)

Outcome: churn proxy.
Predictors: action adoption indicators + scaled tenure since signup + one-hot channel.

This is still not causal; it is a controlled association with included covariates only.


````python
ids = sorted(subscriber_ids)
y = np.array([1 if cid in churned_ids else 0 for cid in ids], dtype=int)

X_act = np.zeros((len(ids), len(actions)), dtype=float)
tenure = np.zeros((len(ids), 1), dtype=float)
channels = []

for i, cid in enumerate(ids):
    tenure[i, 0] = (SNAPSHOT_DATE - signup_by_id[cid]).days
    channels.append([channel_by_id[cid]])
    for j, act in enumerate(actions):
        X_act[i, j] = 1.0 if usage_sum.get((cid, act), 0) > 0 else 0.0

enc = OneHotEncoder(handle_unknown="ignore", sparse_output=False).fit(channels)
X_ch = enc.transform(channels)

tenure_scaled = StandardScaler().fit_transform(tenure)
X = np.concatenate([X_act, tenure_scaled, X_ch], axis=1)

clf = LogisticRegression(
    penalty="l2",
    C=1.0,
    solver="lbfgs",
    max_iter=5000,
    class_weight="balanced",
)
clf.fit(X, y)

proba = clf.predict_proba(X)[:, 1]
print("train_auc", float(roc_auc_score(y, proba)))

coef = clf.coef_.ravel()
odds = np.exp(coef[: len(actions)])

print("\nAdjusted odds ratios for action adoption (proxy outcome=churned):")
for a, o in zip(actions, odds):
    print(f"action_{a}: OR={o:.3f}")
````
